```python
import numpy as np
import pandas as pd
```

```python
origin = pd.read_excel('Gender_StatsEXCEL.xlsx', sheet_name = 'Data')
origin = pd.DataFrame(origin)
origin.head()
```

```python
np.shape(origin)
```

```python
groupby_indicator_names = origin.groupby('Indicator Name')
# np.shape(groupby_indicator_names)
indicator_countries = pd.DataFrame(columns = ['number of countries'])
for group, frame in groupby_indicator_names:
    indicator_countries.loc[group] =  len(frame)   
    
indicator_countries.sort_values(by = 'number of countries')
```

```python
# we want to analyze the data from recent years, try to find some valuable indicators
origin_copy = origin.copy()
origin_copy = origin[['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']]
year = [ '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
origin_copy = origin_copy.dropna(subset = year, how = 'all')
# len(origin_copy['Country Name'].unique()) # 263
# len(origin_copy['Indicator Name'].unique()) #631
# np.shape(origin_copy)
indicator_countries = pd.DataFrame(columns = ['number of countries'])
for group, frame in origin_copy.groupby('Indicator Name'):
    indicator_countries.loc[group] =  len(frame)   
indicator_countries.sort_values(by = 'number of countries', ascending = False)
```

```python
interest_indicator = indicator_countries[indicator_countries['number of countries'] >= 195]
# interest_indicator['Indicator Name'] = interest_indicator.index
interest_indicator.reset_index(level = 0, inplace = True)
interest_indicator.rename(columns = {'index':'Indicator Name'}, inplace = True)
interest_indicator
```

```python
merge = pd.merge(origin_copy, interest_indicator, on = ['Indicator Name'], how = 'right')
merge
```

```python
merge['Indicator Name'].unique()
```

```python
merge.to_csv('159 indicators for all the countries 2000 to 2016.csv')
```
