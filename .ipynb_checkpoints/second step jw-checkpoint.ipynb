```python
region = pd.read_csv('159 indicators for all the countries 2000 to 2016.csv')
eap = region[(region['Country Name'] == 'China') | (region['Country Name'] == 'Japan') | 
               (region['Country Name'] == 'Vietnam') |  (region['Country Name'] == 'Korea, Dem. People\'s Rep.') | 
               (region['Country Name'] == 'Korea, Rep.') | (region['Country Name'] == 'Philippines') | 
               (region['Country Name'] == 'Australia') | (region['Country Name'] == 'New Zealand') | 
               (region['Country Name'] == 'Singapore')]
# region_countries = ['China', 'Japan', 'Vietnam', ]
eap.drop(columns = {'Unnamed: 0', 'Country Code', 'Indicator Code'}, inplace = True)
eap.sort_values(by = 'Country Name', inplace = True)
# eap.set_index('Indicator Name', inplace = True)
eap.append(East_Asia_Pacific)
```

```python
# eap['gender'] = []
def uni_gender(data):
    for index, row in data.iterrows():
        if ', male ' in row['Indicator Name']:
            row['Indicator Name'].replace(', male ', '')
            data.set_value(index, 'Gender', 'M')
        elif ', female ' in row['Indicator Name']:
            row['Indicator Name'].replace(', female ', '')
            data.set_value(index, 'Gender', 'F')
        else:
            data.set_
    return data
        
eap = uni_gender(eap)
eap
        
```
