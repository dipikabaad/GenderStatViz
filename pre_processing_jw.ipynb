```python
import numpy as np
import pandas as pd
```

```python
origin = pd.read_excel('Gender_StatsEXCEL.xlsx', sheet_name = 'Data')
origin = pd.DataFrame(origin)
origin.head()
```

```python
np.shape(origin)
```

```python
groupby_indicator_names = origin.groupby('Indicator Name')
# np.shape(groupby_indicator_names)
indicator_countries = pd.DataFrame(columns = ['number of countries'])
for group, frame in groupby_indicator_names:
    indicator_countries.loc[group] =  len(frame)   
    
indicator_countries.sort_values(by = 'number of countries')
```
